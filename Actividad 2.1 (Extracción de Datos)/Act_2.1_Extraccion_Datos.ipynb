```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CARGA DESDE UN ARCHIVO .XLSX SIN INDICE
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df 
```

FILTRO POR OBJETO

Variables:

  ° Nombre_Cliente: Se eligió esta variable para hacer una busqueda sobre un cliente en especifico que fue Revuelta Montecarlo

```python
filtro1 = df[(df["NOMBRE_CLIENTE"] == "REVUELTA MONTECARLO")]
filtro1
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro1.to_csv("Filtro1_Objeto.csv")
```

FILTRO POR FILAS

Variables:

  ° Filas consecutivas: Se eligieron las primeras 21 filas porque la fecha_doc era la misma

  ° Filas no consecutivas: Se eligieron las filas 29 y 30 porque tienen el mismo nombre del vendedor

```python
#FILAS CONSECUTIVAS
filtro2 = df.iloc[0:20,:] 

#FILAS NO CONSECUTIVAS
#filtro2 = df.iloc[[29,30], :] 
filtro2
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro2.to_csv("Filtro2_Filas.csv")
```

FILTRO POR COLUMNAS

Variables:

  ° Columnas consecutivas: Se eligieron las columnas de la 2 a la 8 porque muestran la información mas concreta 

  ° COlumnas no consecutivas: Se eligieron las columnas 2 y 4 para conocer la relación del vendedor con el numero de articulos vendidos

```python
#COLUMNAS CONSECUTIVAS
#filtro3= df.iloc[ : , 2:8]

#COLUMNAS NO CONSECUTIVAS
filtro3 = df.iloc[ : , [2,4]]
filtro3
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro3.to_csv("Filtro3_Columnas.csv")
```

FILTRO DE FILAS VS COLUMNAS

Variables:

  ° Indice: Se eligió el nombre del ciente como indice porque se quiere hacer una comparación de algunos clientes con el costo total de los articulos

```python
#PASO 1: Guardar el excel con el cambio de indice 
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=3)

#PASO 2: Aplicamos filtro de filas vs columnas
filtro4 = df1.loc[["REDYPLAST", "CROPLASTIC"], ["COSTO_TOTAL_CALCULADO"]]
filtro4
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro4.to_csv("Filtro4_filasVScolumnas.csv")
```

FILTRO POR COMPARACIÓN

Variables:

  ° Precio unitario: se eligió esta variable para conocer el nombre del vendedor que viendo un articulo en su precio unitario arriba de $20000

```python
filtro7 = df[df["PRECIO_UNITARIO"] > 20000]
filtro7
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro7.to_csv("Filtro7_Comparación.csv")
```

FILTRO CON CONECTORES LÓGICOS (AND)

Variables:

  ° Costo unitario: Se eligió esta variable para saber que productos tienen un costo unitario menor a $100

  ° Elicamex: Se eligió esta variable para conocer si se tenian productos con costos unitarios menores a $100 y que se hayan adquirido de este cliente

```python
filtro8 = df[(df["COSTO_UNITARIO"] < 100) & (df["NOMBRE_CLIENTE"]=="ELICAMEX")]
filtro8
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro8.to_csv("Filtro8_ConectorLogicoAnd.csv")
```

FILTRO CON CONECTORES LÓGICOS (O)

Variables:

  ° Precio unitario: Se eligió esta variable para saber que productos tienen un precio menor a $1000

  ° Fecha: Se eligió esta variable para saber que productos se adquieron el 30 de Junio o que tengan un precio unitario menor a $1000

```python
filtro9 = df [(df["PRECIO_UNITARIO"] < 6000) | (df["FECHA_DOC"] == "2022-06-30")]
filtro9
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro9.to_csv("Filtro9_ConectorLogicoOr.csv")
```

FILTRO CON CONECTORES LOGICOS (NOT)

Variables:

  ° Nombre cliente: se eligió esta variable para quitar los datos cuando el cliente es "Público en general" 

```python
filtro10 = df[~ (df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro10
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro10.to_csv("Filtro10_ConectorLogicoNot.csv")
```
