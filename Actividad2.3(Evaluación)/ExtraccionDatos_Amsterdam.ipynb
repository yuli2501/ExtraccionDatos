```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CARGA DESDE UN ARCHIVO .XLSX SIN INDICE
df= pd.read_csv('Amsterdam.csv') 
df 
```

FILTRO POR FILAS

Variables:

  ° Filas no consecutivas: Se eligieron las filas con el host name de Edwin, Carolien, Jan, Flip y Karin, para conocer los datos solo de estas filas

```python
#FILAS NO CONSECUTIVAS
filtro1 = df.iloc[[2,9654,4,1,7], :] 
filtro1
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro1.to_csv("Filtro1(Filas).csv") 
```

FILTRO POR COMPARACIÓN

Variables:

  ° Last_review: se eligió esta variable para conocer los host que se unieron antes del 2019

```python
filtro2 = df[pd.to_datetime(df["host_since"]).dt.year < 2019]
filtro2
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro2.to_csv("Filtro2(Comparación).csv") 
```

FILTRO POR OBJETO

Variables:

  ° host_response_time: se eligió esa variable para saber que host tardan por mucho un dia en responder

```python
filtro3 = df[df['host_response_time'] == 'within a day']
filtro3

```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro3.to_csv("Filtro3(Objeto).csv") 
```

FILTRO COMPARACIÓN

Variables:

  ° Accomodates: Se eligió esta variable para buscar los valores que tienen al menos 5 


```python
filtro4 = df[df['accommodates'] >= 5]
filtro4
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro4.to_csv("Filtro4(Comparacion).csv") 
```

FILTRO POR COLUMNAS

Variables:

  ° Columnas no consecutivas: Se eligieron las columnas "CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT" para conocer la relaciónque existe entre las fechas con el precio total


```python
filtro5 = df[df['room_type'].isin(['Entire home/apt', 'Private room'])]
filtro5
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro5.to_csv("Filtro5.csv") 
```
