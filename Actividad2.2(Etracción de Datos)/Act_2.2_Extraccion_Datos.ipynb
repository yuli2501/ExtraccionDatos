```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CARGA DESDE UN ARCHIVO .XLSX SIN INDICE
df= pd.read_excel('Datos de Facturación.xlsx')
df 
```

FILTRO CON CONECTORES LÓGICOS (AND)

Variables:

  ° CVE_CLPV: Se eligió esta variable para conocer los datos que van de 1000 hasta los 2000

```python
df["CVE_CLPV"] = pd.to_numeric(df["CVE_CLPV"], errors="coerce")

filtro1 = df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro1.to_csv("Filtro1(ConectorLogicoAnd).csv") 
```

FILTRO CON CONECTORES LOGICOS (NOT)

Variables:

  ° CVE_VEND: se eligió esta variable para quitar las claves 5 y 4

```python
filtro2 = df[~ (df["CVE_VEND"] == 4 ) & ~ (df["CVE_VEND"] == 5 )]
filtro2
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro2.to_csv("Filtro2(ConectorLogicoNot).csv") 
```

FILTRO POR OBJETO

Variables:

  ° FECHA_ENT: Se eligió esta variable para hacer una busqueda sobre una fecha en especifico que fue "28/02/2022"

```python
# Convertir la columna a datetime
df["FECHA_ENT"] = pd.to_datetime(df["FECHA_ENT"], format="%d/%m/%Y %I:%M:%S %p", errors="coerce")

# Filtrar por fecha exacta
filtro3 = df[df["FECHA_ENT"].dt.date == pd.to_datetime("2022-02-28").date()]
filtro3

```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro3.to_csv("Filtro3(Objeto).csv") 
```

FILTRO CON CONECTORES LÓGICOS (O)

Variables:

  ° CANT_TOT: Se eligió esta variable para buscar los valores que tienen una cantidad menor a 5951.7 

  ° STATUS: Se eligió esta variable para buscar los valores que tuvieran un estatus tipo "E" o una cantidad menor a 5951.7


```python
#FILTRO BASICO CON CONECTORES LOGICOS (O)
filtro4 = df [(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro4.to_csv("Filtro4(ConectorLogicoOr).csv") 
```

FILTRO POR COLUMNAS

Variables:

  ° Columnas no consecutivas: Se eligieron las columnas "CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT" para conocer la relaciónque existe entre las fechas con el precio total


```python
#COLUMNAS NO CONSECUTIVAS
filtro5 = df.iloc[ : , [0,6,7,9]]
filtro5
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro5.to_csv("Filtro5(Columnas).csv") 
```

FILTRO POR FILAS

Variables:

  ° Filas consecutivas: Se eligieron las filas que van de 7001 a 7099 para conocer los datos solo de estas filas


```python
#FILAS CONSECUTIVAS
filtro6 = df.iloc[7001:7100,:] 
filtro6
```

```python
#CONVERTIR ARCHIVO FILTRADO A CSV
filtro6.to_csv("Filtro6(Filas).csv") 
```

FILTRO DE FILAS VS COLUMNAS

Variables:

  ° Indice: Se eligió el CVE_VEND como indice porque se quieren obtener los datos de dicha variable cuando sus valores son 1 y 2, además de la fecha de elaboración 


```python
#PASO 1: Guardar el excel con el cambio de indice 
df1 = pd.read_excel('Datos de Facturación.xlsx', index_col=3)

#PASO 2: Aplicamos filtro de filas vs columnas
filtro7 = df1.loc[[1,2], ["FECHAELAB"]]
filtro7
```
